```python
def normalize_pwm(pwm, factor=None, max=None):
    if not max:
        max = np.max(np.abs(pwm))
    pwm = pwm/max
    if factor:
        pwm = np.exp(pwm*factor)
    norm = np.outer(np.ones(pwm.shape[0]), np.sum(np.abs(pwm), axis=0))
    pwm = pwm/norm
    pwm[np.isnan(pwm)] = 0.25 # background
    return pwm
```
