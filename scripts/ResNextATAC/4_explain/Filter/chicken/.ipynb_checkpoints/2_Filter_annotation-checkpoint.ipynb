```python
d = {}
for _, gene in tomtom_anno[["Query_ID", "gene_name"]].iterrows():
    m = gene.Query_ID
    tf = gene.gene_name
    if m not in d.keys():
        d[m] = [tf]
    else:
        tfs = d[m]
        if tf not in tfs:
            d[m].append(tf)

#d["Motif_58"]
```

```python
for k,v in d.items():
    d[k] = '; '.join(v)

#d["Motif_58"]
```

```python
filter_tf_anno = pd.DataFrame(d, index=['tfs']).T
#filter_tf_anno.to_csv("./tomtom_conv1_Cistarget/filter_tf_anno.csv")
filter_tf_anno
```
